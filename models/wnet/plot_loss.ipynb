```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
# wnet_1 = "./1_class/ctest_wnet_1class.txt"
wnet_2 = "./2_class/test_wnet_2class.txt"
```

```python
def plot(loss_record_path):
    with open(loss_record_path, 'rb') as f:
        ncuts, rec = pickle.load(f)
    max_ncuts = np.max(ncuts)
    max_rec = np.max(rec)

    ncuts = ncuts / max_ncuts
    rec = rec / max_rec

    plt.xlabel('Epochs')
    plt.ylabel('Normalized loss')
    plt.title(f'WNet loss record : {loss_record_path}')
    plt.plot(ncuts)
    plt.plot(rec)
    plt.legend(['Soft Ncuts', 'Reconstruction'])
    plt.show()
```

```python
for path in [wnet_2]:
    plot(path)
```
