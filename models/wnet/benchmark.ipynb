```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('WNet_results/1500e_loss_record.txt', 'rb') as f:
    ncuts, rec = pickle.load(f)

```

```python
max_ncuts = np.max(ncuts)
max_rec = np.max(rec)

ncuts = ncuts / max_ncuts
rec = rec / max_rec

plt.xlabel('Epochs')
plt.ylabel('Normalized loss')
plt.title('WNet loss record')
plt.plot(ncuts)
plt.plot(rec)
plt.legend(['Soft Ncuts', 'Reconstruction'])
plt.show()
```
